```python
X = df[[f'var_{i}' for i in range(1, 81)]]
y = df['yield']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()

X_train_proc = scaler.fit_transform(X_train)
X_test_proc = scaler.transform(X_test)
n = len(y_test)
```

```python
from scipy.cluster.hierarchy import linkage, cut_tree
from scipy.spatial.distance import pdist
from sklearn.metrics import silhouette_score

iris_e = pdist(X_train_proc, metric='euclidean')
iris_es = linkage(iris_e, method='single')

cluster_es = cut_tree(iris_es, n_clusters=4).flatten()

silhouette_avg = silhouette_score(X_train_proc, cluster_es)

print("For n_clusters = 4, The average silhouette_score is :", silhouette_avg)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.cm as cm
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris, make_blobs, make_moons
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from matplotlib.colors import Normalize
from matplotlib.patches import Ellipse
import statsmodels.api as sm
import warnings
import matplotlib.cm as cm
warnings.filterwarnings("ignore")
```

```python
kmeans_result = KMeans(n_clusters=4, random_state=42).fit(X_train_proc)
cluster_labels = kmeans_result.fit_predict(X_train_proc)
silhouette_avg = silhouette_score(X_train_proc, cluster_labels)
print("For n_clusters =", 4, "The average silhouette_score is :", silhouette_avg)
```

```python
gmm = GaussianMixture(n_components=4, covariance_type='full', random_state=42)
gmm.fit(X_train_proc)

# Predict cluster assignments and probabilities
clusters = gmm.predict(X_train_proc)

silhouette_avg = silhouette_score(X_train_proc, clusters)
print("For n_clusters =", 4, "The average silhouette_score is :", silhouette_avg)
```

```python
from sklearn.metrics import adjusted_rand_score

# Compute Adjusted Rand Index (ARI) for both methods
ari_kmeans = adjusted_rand_score(y_train, cluster_labels)
ari_gmm = adjusted_rand_score(y_train, clusters)

# Display ARI comparison
ari_comparison = pd.DataFrame({
    "Clustering Method": ["K-Means", "Gaussian Mixture Model (GMM)"],
    "ARI Score": [ari_kmeans, ari_gmm]
})

ari_comparison
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_validate

X = df[[f'var_{i}' for i in range(1, 81)]]
y = df['yield']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()

ridge_pipeline = Pipeline([
    ('scaler', scaler),
    ('regressor', Ridge(alpha=1))  # Adjust alpha as desired
])

qda_pipeline = Pipeline([
    ('scaler', scaler),
    ('regressor', QuadraticDiscriminantAnalysis)
])


ridge_pipeline.fit(X_train_proc, ctrain)
y_pred_ridge = ridge_pipeline.predict(X_train_proc)

qda_iris = QuadraticDiscriminantAnalysis()
qda_iris.fit(X_train, ctrain)
y_pred_qda = qda_iris.predict(X_train)
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error, r2_score, accuracy_score
from sklearn.model_selection import cross_val_score, KFold, LeaveOneOut
```

```python
# Define scoring metrics
scoring = {
    'accuracy': make_scorer(accuracy_score)
}

# K-Fold Cross-Validation
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=123)

print(f"\n {k}-Fold Cross-Validation:")
kfold_results = {}
for metric, scorer in scoring.items():
    scores = cross_val_score(qda_pipeline, X_train, y_pred_qda, scoring=scorer, cv=kf)
    kfold_results[metric] = (np.mean(scores), np.std(scores))
    print(f"{metric}: Mean = {scores.mean():.3f}, Std = {scores.std():.3f}")
```
